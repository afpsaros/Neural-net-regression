```python
from __future__ import print_function

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

import time
```

```python
# tf.random.set_random_seed(seed = 5)
x = {'key': tf.Variable(tf.random_normal([1, 1]))}
obj = tf.square(x['key'])
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.2)
train_op = optimizer.minimize(obj)
```

```python
sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)
print(sess.run(x))
for i in range(10):
    sess.run(train_op)
    print(sess.run(x))
    
x = {'key': tf.Variable(sess.run(x)['key'])}
sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)

print('repeat', sess.run(x))

obj = tf.square(x['key'])
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.2)
train_op = optimizer.minimize(obj)
for i in range(10):
    sess.run(train_op)
    print(sess.run(x))
```
