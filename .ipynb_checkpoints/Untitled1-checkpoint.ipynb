```python
import tensorflow as tf
import numpy as np
import time

class DNN:
    def __init__(self, X, layer_size, Xmin, Xmax):
        self.X = X
        self.size = layer_size
        self.Xmin = Xmin
        self.Xmax = Xmax
    
    def hyper_initial(self):
        L = len(self.size)
        W = []
        b = []
        for l in range(1, L):
            in_dim = self.size[l-1]
            out_dim = self.size[l]
            std = np.sqrt(2/(in_dim + out_dim))
            weight = tf.Variable(tf.random_normal(shape=[in_dim, out_dim], stddev=std))
            bias = tf.Variable(tf.zeros(shape=[1, out_dim]))
            W.append(weight)
            b.append(bias)

        return W, b

    def fnn(self, W, b):
        A = 2.0*(self.X - self.Xmin)/(self.Xmax - self.Xmin) - 1.0
        L = len(W)
        for i in range(L-1):
            A = tf.tanh(tf.add(tf.matmul(A, W[i]), b[i]))
        Y = tf.add(tf.matmul(A, W[-1]), b[-1])
        
        return Y
```

```python
import matplotlib.pyplot as plt

# from net import DNN 

np.random.seed(1234)
tf.set_random_seed(1234)
 

layers = [1] + 3*[20] + [1]
x = np.linspace(-1, 2, 21).reshape((-1, 1))
y = x**2

Xmin = x.min(0)
Xmax = x.max(0)

x_train = tf.placeholder(shape=[None, 1], dtype=tf.float32)
y_train = tf.placeholder(shape=[None, 1], dtype=tf.float32)

model = DNN(x_train, layers, Xmin, Xmax)
W, b = model.hyper_initial()
y_pred = model.fnn(W, b)

loss = tf.reduce_mean(tf.square(y_pred - y_train))
train = tf.train.AdamOptimizer().minimize(loss)

sess = tf.Session()
sess.run(tf.global_variables_initializer())
tr_error = [sess.run(loss, feed_dict={x_train: x, y_train: y})]
nmax = 10000
n = 0
tic = time.perf_counter()
while n < nmax:
    n += 1
    y_, loss_, _ = sess.run([y_pred, loss, train], feed_dict={x_train: x, y_train: y})
    tr_error.append(loss_)
#     if n%100 == 0:
#         print('n: %d, loss: %.3e'%(n, loss_))
toc = time.perf_counter()
x_test = np.linspace(-1, 2, 1001).reshape((-1, 1))
y_test = sess.run(y_pred, feed_dict={x_train: x_test})
#     plt.figure()
plt.yscale('log')
plt.plot(tr_error)
plt.show()
print(toc-tic)
```
