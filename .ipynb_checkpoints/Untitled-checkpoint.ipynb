```python
from __future__ import print_function

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

import time
```

```python
n = 30
X_data = np.linspace(-1, 2, n)
X_data = X_data.reshape((n, 1))
# Y_data = X_data**2
Y_data = np.cos(X_data / 5)**3 + 4* np.sin(2 * X_data)**3 + \
.3 * (X_data - 5)**2 + 0.02 * (X_data - 2)**3 

Y_data += np.random.normal(0, 0.1, [n, 1])

plt.plot(X_data, Y_data, 'bo')
plt.show()

data = np.concatenate((X_data, Y_data), axis=1)

np.random.shuffle(data)

data_tr, data_val= np.split(data, [int(0.8*n)], axis = 0)
print(data_tr.shape, data_val.shape)
n_tr, n_val = data_tr.shape[0], data_val.shape[0]

n_eval = 300
X_data = np.linspace(-1, 2, n_eval)
X_data = X_data.reshape((n_eval, 1))
# Y_data = X_data**2
Y_data = np.cos(X_data / 5)**3 + 4* np.sin(2 * X_data)**3 + \
.3 * (X_data - 5)**2 + 0.02 * (X_data - 2)**3 

# plt.plot(X_data, Y_data, 'bo')
# plt.show()

data_eval = np.concatenate((X_data, Y_data), axis=1)
```

```python
def next_batch(data_tr, i, batch_size):
    return data_tr[i * batch_size:(i + 1) * batch_size - 1, :] 
```

```python
# various parameters
num_epochs = 10000
batches = 1
batch_size = n_tr // batches
itpep = n_tr // batch_size

wd_par = 0
lr = 1e-2

num_input = [1] 
num_output = [1] 

layers = 4
width = 100
widths = num_input + layers * [width] + num_output

w_sizes = [[widths[i], widths[i+1]] for i in range(len(widths) - 1)]
w_keys = [('').join(('h', str(i+1))) for i in range(layers)] + ['out']

w_vars = []
for s in w_sizes:
    std = np.sqrt(2/(s[0] + s[1]))
    w_vars.append(tf.Variable(tf.random_normal(s, stddev = std)))

# w_vars = [tf.Variable(tf.random_normal(s)) for s in w_sizes]
weights = dict(zip(w_keys, w_vars))

b_sizes = widths[1:]
b_keys = [('').join(('b', str(i+1))) for i in range(layers)] + ['out']
b_vars = [tf.Variable(tf.zeros([s])) for s in b_sizes]
biases = dict(zip(b_keys, b_vars))

# tf Graph input
X = tf.placeholder(tf.float32, shape = [None, 1])
Y = tf.placeholder(tf.float32, shape = [None, 1])

# create model
Xmin = -1
Xmax = 2
def ffn(x):
    A = 2.0*(x - Xmin)/(Xmax - Xmin) - 1.0
    layer = tf.tanh(tf.add(tf.matmul(A, weights['h1']), biases['b1']))    
    for i in range(1, layers):
        layer = tf.tanh(tf.add(tf.matmul(layer, weights[w_keys[i]]), biases[b_keys[i]]))        
    layer = tf.add(tf.matmul(layer, weights['out']), biases['out'])
    return layer

fp = ffn(X)
error = tf.reduce_mean(tf.square(fp - Y))
error_p = tf.math.abs(fp - Y)
obj = error + wd_par * tf.reduce_sum([tf.reduce_sum(tf.square(weights[i])) for i in weights])

optimizer = tf.train.AdamOptimizer()
train_op = optimizer.minimize(obj)
init = tf.global_variables_initializer()
```

```python
sess = tf.Session()
sess.run(init)
tr_error = [sess.run(error, feed_dict={X: data_tr[:, [0]], Y: data_tr[:, [1]]})]
val_error = [sess.run(error, feed_dict={X: data_val[:, [0]], Y: data_val[:, [1]]})]
check_w = [sess.run(weights)]
check_b = [sess.run(biases)]

pat = 10000
v = 1e10
j = 0
best = -1
tic = time.perf_counter()
for epoch in range(num_epochs):
    if itpep == 1:
        tr_error_, weights_, biases_, _ = \
        sess.run([error, weights, biases, train_op], feed_dict={X: data_tr[:, [0]], Y: data_tr[:, [1]]})
        check_w.append(weights_)
        check_b.append(biases_)
        tr_error.append(tr_error_)
        val_error.append(sess.run(error, feed_dict={X: data_val[:, [0]], Y: data_val[:, [1]]}))       
    else: 
        for i in range(itpep):
            data_batch = next_batch(data_tr, i, batch_size)
            sess.run(train_op, feed_dict={X: data_batch[:, [0]], Y: data_batch[:, [1]]})
        tr_error_, weights_, biases_= \
        sess.run([error, weights, biases], feed_dict={X: data_batch[:, [0]], Y: data_batch[:, [1]]})
        check_w.append(weights_)
        check_b.append(biases_)
        tr_error.append(tr_error_)
        val_error.append(sess.run(error, feed_dict={X: data_val[:, [0]], Y: data_val[:, [1]]}))   
        
    if epoch > 2000:
        if val_error[-1] < v:
            j = 0
            best = epoch
            best_w = check_w[-1]
            best_b = check_b[-1]
            v = val_error[-1]
        else:
                j += 1
        if j == pat:
            break        
toc = time.perf_counter()
print(toc-tic)
print('Optimization finished with validation RMSE {:.6f}'.format(np.sqrt(val_error[-1])))
print('{:03d} epochs run'.format(epoch + 1))
print('best parameters at epoch {:03d}'.format(best + 1))

plt.title('Training vs validation error', fontsize = 20)
plt.yscale('log')
plt.axvline(x = best, label = 'selected epoch', color = 'r')
plt.plot(tr_error, label = 'training')
plt.plot(val_error, label = 'validation')
plt.xlabel('Epoch', fontsize = 15)
plt.ylabel('Error', fontsize = 15)
plt.legend(bbox_to_anchor=(1.4, 1.0))
plt.show() 
```
